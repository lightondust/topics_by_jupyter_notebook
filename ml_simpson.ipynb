```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = np.array([0,0,0])
pre = np.array([
    [1.0, 0., 0.5],
    [0., 1.0, 0.6],
    [0.5, 0.6, 1.0]
])
```

```python
cov = np.linalg.inv(pre)
```

```python
samples = np.random.multivariate_normal(mean=mean, cov=cov, size=10000)
```

```python
samples.shape
```

```python
x = samples[:,0]
y = samples[:,1]
z = samples[:,2]
```
