```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def get_rand_norm_vec(dim, n_sample=200000):
    rand_vec = np.random.rand(n_sample, dim) * 2 - 1
    norm = np.sqrt((rand_vec**2).sum(axis=1))
    rand_norm_vec = rand_vec / norm.reshape(-1, 1)
    return rand_norm_vec
```

```python
n_sample = 200000
dim = 100
rand_vec = np.random.rand(n_sample, dim) * 2 - 1
norm = np.sqrt((rand_vec**2).sum(axis=1))
rand_norm_vec = rand_vec / norm.reshape(-1, 1)
```

```python
plt.hist(rand_vec[:,0])
```

```python
plt.hist(rand_norm_vec[:, 2], range=(-1, 1), bins=40)
```

```python
rand_norm_vec.max(axis=0)
```

```python
abs(rand_norm_vec).mean(axis=0)
```

```python
prob_list = []
dim_list = [d for d in range(1, 200, 10)]
for d in dim_list:
    prob_list.append(abs(get_rand_norm_vec(d)).mean())
```

```python
plt.figure(figsize=(12, 9))
plt.plot(dim_list, prob_list)
```
