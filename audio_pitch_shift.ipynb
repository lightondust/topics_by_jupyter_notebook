```python
import librosa
import numpy as np
from pydub import AudioSegment
```

```python
default_rate = 44100
```

```python
def numpy_to_audio_segment(audio_arr, amp=1.0, rate=default_rate, channels=1, sample_width=4):
    audio_arr = audio_arr / (abs(audio_arr).max())
    audio_arr_trans = ((2**(8*sample_width-1)-1) * amp * audio_arr).astype(np.int32)
    audio_segment = AudioSegment(audio_arr_trans.tobytes(),
                                 frame_rate=rate,
                                 sample_width=sample_width,
                                 channels=channels)
    return audio_segment
```

```python
wav_path = '/home/sen/Music/sound_a.wav'
```

```python
audio_arr, sr = librosa.load(wav_path, sr=None)
```

```python
numpy_to_audio_segment(audio_arr, rate=sr)
```

```python
numpy_to_audio_segment(librosa.effects.pitch_shift(audio_arr, sr=sr, n_steps=3), rate=sr)
```

```python
numpy_to_audio_segment(librosa.effects.pitch_shift(audio_arr, sr=sr, n_steps=7), rate=sr)
```
